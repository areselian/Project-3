```python
import pandas as pd
```

```python
DF1 = pd.read_csv("NBAdataset_new.csv")
DF2 = pd.read_csv("Coords.csv")
```

```python
DF1_ = pd.DataFrame(DF1)
DF1_1 = DF1_.drop(columns=['Unnamed: 0'])
DF1_1.head()
```

```python
DF2_ = pd.DataFrame(DF2)
DF2_2 = DF2_.rename(columns={"name":"college"})
DF2_2.head()
```

```python
Merged = pd.merge(DF1_1, DF2_2, how='right', on='college')
Merged.head(40)
```

```python
Merged.to_csv('LatLongNBA.csv', index=False)
```
